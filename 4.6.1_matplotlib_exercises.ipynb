```python
import matplotlib.pyplot as plt
import math
import numpy as np
%matplotlib inline
```

```python
print(plt)
```

```python
x = list(range(-10, 10))
y = [(n ** 2 - n + 2) for n in x]
plt.plot(x,y)
plt.annotate('Origin', xy=(0, 0), xytext=(-3, 20),
             arrowprops={'facecolor': 'red'})
```

```python
x = list(range(0, 100))
y = [(n ** (1/2) ) for n in x]
plt.xlim(0, 100)
plt.ylim(0, 12)
plt.plot(x,y)
```

```python
x = list(range(-20, 20))
y = [(n ** (3) ) for n in x]
plt.xlim(-20, 20)
plt.ylim(-8000, 8000)
plt.plot(x,y)
```

```python
x = list(np.arange(-10, 10,.01))
y = [math.tan(n) for n in x]
plt.xlim(-2, 2)
plt.ylim(-6, 6)
plt.plot(x, y)
```

```python
x = list(range(0, 10))
y = [(2 ** n) for n in x]
plt.xlim(0, 10)
plt.ylim(0, 600)
plt.plot(x,y)
```

```python
plt.subplot(221)
x = list(range(0, 100))
y = [(n ** (1/2) ) for n in x]
plt.xlim(0, 100)
plt.ylim(0, 12)
plt.plot(x, y)
plt.title('$y = \sqrt{x}$')
plt.tight_layout()
plt.subplot(222)
x = list(range(-20, 20))
y = [(n ** (3) ) for n in x]
plt.xlim(-20, 20)
plt.ylim(-8000, 8000)
plt.plot(x, y)
plt.title('$y = x^3$')
plt.tight_layout()
plt.subplot(223)
x = list(np.arange(-10, 10,.01))
y = [math.tan(n) for n in x]
plt.xlim(-2, 2)
plt.ylim(-6, 6)
plt.plot(x, y)
plt.title('$y = tan(x)$')
plt.tight_layout()
plt.subplot(224)
x = list(range(0, 10))
y = [(2 ** n) for n in x]
plt.xlim(0, 10)
plt.ylim(0, 600)
plt.plot(x, y)
plt.title('$y = 2^x$')
plt.tight_layout()
```

```python
x = list(range(0, 100))
y = [(n ** (1/2) ) for n in x]
plt.plot(x,y, label = '$\sqrt{x}$')
x = list(range(-20, 20))
y = [(n ** (3) ) for n in x]
plt.plot(x,y, label = '$x^3$')
x = list(np.arange(-10, 10,.01))
y = [math.tan(n) for n in x]
plt.plot(x, y, label = '$tan(x)$')
x = list(range(0, 10))
y = [(2 ** n) for n in x]
plt.xlim(0, 10)
plt.ylim(0, 100)
plt.plot(x,y, label = '$2^x$')
plt.title('Ugly Chart')
plt.legend(loc='upper right')
```
